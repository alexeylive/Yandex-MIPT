#### All imported libraries are here

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

#### I will use ARGUS distribution(https://en.wikipedia.org/wiki/ARGUS_distribution)

```python
chi = 1
argus_rv = sts.argus(chi)
samples = argus_rv.rvs(1000)
```

#### Sample histogram and theoretical probability density

```python
x = np.linspace(0,1, 1000)
pdf = argus_rv.pdf(x)
plt.hist(samples,density = True,label = 'Practical')
plt.plot(x, pdf,'r', label='Theoretical');
plt.legend()
plt.xlabel('x');
plt.ylabel('f(x)');
```

#### Functions for plotting probability density of sample means

```python
def create_graph(distribution,n):
    create_hist(distribution.rvs,n)
    mean, variance = [x + 0 for x in distribution.stats()]
    x = np.linspace(0,1, 1000)
    plt.plot(x, sts.norm.pdf(x, mean, np.sqrt(variance/n)),'r',label = 'Theoretical')
    plt.legend()
    plt.title('Probability density of sample means, n = ' + str(n));
    plt.xlabel('x');
    plt.ylabel('f(x)');
    
```

```python
def create_hist(rvs,n):
    samp_mean = []
    for i in range(0,1000):
        rv = rvs(n)
        samp_mean += [np.mean(rv)]
    plt.hist(samp_mean,density = True,label = 'Practical')
```

#### Formation of distributions of sample means


```python
create_graph(argus_rv,5)
```

```python
create_graph(argus_rv,10)
```

```python
create_graph(argus_rv,50)
```

### The main postulates of CLT are fulfilled:
* The histogram of the distribution of sample means can be described by the normal distribution with the parameters Ex, $\frac{Dx}{n}$
* As the sample size increases, the histogram of the sample means distribution begins to resemble the normal distribution more accurately, and the variance decreases.
